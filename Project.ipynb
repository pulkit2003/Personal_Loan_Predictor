```python
file = pd.read_csv('Loan_Modelling.csv', index_col=0)
```

```python
file.columns
```

```python
x,y = file.drop(['Age', 'Experience','ZIPCode','Family','Securities_Account','Online', 'CreditCard','Personal_Loan'], axis=1) , file['Personal_Loan']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
```

```python
ypred = model.predict(xtest)
```

```python
app = tkn.Tk()
app.geometry('1000x600')
app.title('Personal Loan Predictor')
app.configure(bg='cyan')
app.config(cursor= "")
img = PhotoImage(file = "Bank.png")
label = Label(app,image=img).place(x=0,y=0)
mod = pd.read_pickle('Personal_Loan_Customers.pkl')

inc = tkn.Variable(app)
ccavg = tkn.Variable(app)
edu = tkn.Variable(app)
mort = tkn.Variable(app)
cd_acc=tkn.Variable(app)
result=tkn.Variable(app)

tkn.Label(app, text="Personal Loan Prediction",font=('Times New Roman',20),foreground='cyan', bg='black').place(x=300,y=30)
tkn.Label(app, text="Income :     ",font=('Arial',15),foreground='cyan', bg='black').place(x=180,y=100)
tkn.Label(app, text="CC Average :  ",font=('Arial',15),foreground='cyan', bg='black').place(x=180,y=160)
tkn.Label(app, text="Education Qualification:      ",font=('Arial',15),foreground='cyan', bg='black').place(x=180,y=220)
tkn.Label(app, text="Mortgage Value : ",font=('Arial',15),foreground='cyan', bg='black').place(x=180,y=320)
tkn.Label(app, text="Certificate of Deposit:  ",font=('Arial',15),foreground='cyan', bg='black').place(x=180,y=380)
tkn.Label(app, textvariable=result,font=('Arial',20),fg='cyan', bg='black').place(x=480,y=540)
    
tkn.Entry(app,textvariable=inc, width=12, font=('Arial',15),foreground='cyan', bg='black', insertbackground="red").place(x=520,y=100)
tkn.Entry(app,textvariable=ccavg, width=12, font=('Arial',15),foreground='cyan', bg='black', insertbackground="red").place(x=520,y=160)
tkn.Entry(app,textvariable=edu, width=12, font=('Arial',15),foreground='cyan', bg='black', insertbackground="red").place(x=520,y=220)
tkn.Entry(app,textvariable=mort, width=12, font=('Arial',15),foreground='cyan', bg='black', insertbackground="red").place(x=520,y=320)
tkn.Entry(app,textvariable=cd_acc, width=1, font=('Arial',15),foreground='cyan', bg='black', insertbackground="red").place(x=660,y=380)

def ed_u():
    edu.set('1')
def ed_g():
    edu.set('2')
def ed_p():
    edu.set('3')
    
def cd_y():
    cd_acc.set('1')
def cd_n():
    cd_acc.set('0')
    
def prediction():
    global mod
    
    if (edu==1):
        ed_u()
    elif(edu==2):
        ed_g()
    elif(edu==3):
        ed_p()
        
    if (cd_acc==1):
        cd_y()
    elif(cd_acc==0):
        cd_n()
        
    income = eval(inc.get())
    cc_avg = eval(ccavg.get())
    education = eval(edu.get())         
    mortgage = eval(mort.get())
    cd_account = eval(cd_acc.get())
    query= [[income,cc_avg,education,mortgage,cd_account]]
    p= round(m.predict(query)[0],1)
    if(p==1):
         result.set("YES")
    else:
        result.set("NO")

tkn.Button(app, text='The Customer Will Buy A Loan', font=('Arial',15), bg='black',foreground='cyan', pady=10,command=prediction).place(x=320,y=440)

tkn.Button(app, text='Under Graduate', font=('Arial',10), bg='black',foreground='cyan', pady=5,command=ed_u).place(x=280,y=260)
tkn.Button(app, text='Graduate', font=('Arial',10), bg='black',foreground='cyan', pady=5,command=ed_g).place(x=440,y=260)
tkn.Button(app, text='Advance/Professional', font=('Arial',10), bg='black',foreground='cyan', pady=5,command=ed_p).place(x=560,y=260)

tkn.Button(app, text='   Yes   ', font=('Arial',10), bg='black',foreground='cyan', pady=5,command=cd_y).place(x=560,y=380)
tkn.Button(app, text='   No   ', font=('Arial',10), bg='black',foreground='cyan', pady=5,command=cd_n).place(x=660,y=380)

app.mainloop()
```
